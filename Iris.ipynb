```python
#converting the categorical data into numerical data 
le=preprocessing.LabelEncoder()
iris_df['species']=le.fit_transform(iris_df['species'])
```

```python
#Using 80-20, just like Pareto, you know
train, test=train_test_split(iris_df, test_size=0.2, random_state=0)
train.shape
test.shape
```

```python
help(train_test_split)
```

```python
train_x=train.drop('species', axis='columns')
train_x
train_y=train['species']
train_y
```

```python
test_x=test.drop('species', axis='columns')
test_y=test['species']
test_x
```

```python
#Using Logistic Regression because our target variable is categorical.
model=LogisticRegression()
model.fit(train_x, train_y)
```

```python
prediction=model.predict(test_x)
```
